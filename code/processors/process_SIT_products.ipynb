```python
import numpy as np
from netCDF4 import Dataset
from os import listdir
from os.path import isfile, join
import pandas as pd
import pickle
import xarray as xr
import datetime
```

```python
yearlist = np.arange(2011,2021)
monthlist = ["10","11","12","01","02","03","04"]

awi_dir = '/home/robbie/Dropbox/SM_Thickness_Thesis/data/AWI_CS2/'

awi_sit = {}

for year in yearlist:
    
    awi_sit[year] = {}
    
    for month in monthlist:
        
        d = Dataset(f'{awi_dir}awi-siral-l3c-sithick-cryosat2-rep-nh_25km_ease2-{year}{month}-fv2p3.nc')

        sit = np.array(d['sea_ice_thickness'])[0]
        
        awi_sit[year][month] = sit
```

```python

awidir = '../../data/NSIDC/NSIDC/'

onlyfiles = [f for f in listdir(awidir) if isfile(join(awidir, f))]



yearlist = np.arange(2011,2021)

nsidc_sit = {}

for year in yearlist:
    
    nsidc_sit[year] = {}
    
    for month in monthlist:
        
        nsidc_sit[year][month] = {}
        
        if month == '12':
            m = '01'
        else: 
            m = str(int(month)+1).zfill(2)
        
        s = f'{year}{m}'
    
#         print(s)
        
        try:
            file = [f for f in onlyfiles if s in f][0]
        except:
            print(s)

        d = Dataset(f'{awidir}{file}')

        sit = np.array(d['sea_ice_thickness'])
        
        sit[sit>100] = np.nan
        sit[sit<0] = np.nan
        
        nsidc_sit[year][month] = sit
```

```python
rho_w = 1028
rho_i = 900

data_dir = '../../data/W99/'

W99_sit = {}

with xr.open_dataset(f'{data_dir}W99_361.nc') as data:

    for year in yearlist:

        W99_sit[year] = {}

        for month in monthlist:

            ds_month = data.where(int(month) == data.month, drop=True)

            SWE = np.array(ds_month['SWE'])[0]
#             depth = np.array(ds_month['depth'])[0]
            
            SNOW_SIT = SWE * (rho_w/(rho_w-rho_i))*1.81e-3
            
            W99_sit[year][month] = SNOW_SIT
```
